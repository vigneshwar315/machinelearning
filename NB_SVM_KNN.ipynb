```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
dataset=pd.read_csv("/content/diabetes_DT.csv")
X=dataset.drop("Outcome",axis=1)
y=dataset["Outcome"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
model=DecisionTreeClassifier(max_depth=3)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print("accuracy:",accuracy)
plt.figure(figsize=(20,10))
plot_tree(model,feature_names=X.columns,class_names=["diabetes","no diabetes"])
plt.show()
```

```python
#naivebayes
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.preprocessing import LabelEncoder


dataset=pd.read_csv("/content/Loan payments data.csv")
dataset.drop(["Loan_ID","effective_date","due_date","paid_off_time"],axis=1,inplace=True)
dataset=dataset.fillna(0)
le=LabelEncoder()
dataset["loan_status"]=le.fit_transform(dataset["loan_status"])
dataset["education"]=le.fit_transform(dataset["education"])
dataset["Gender"]=le.fit_transform(dataset["Gender"])

X=dataset.drop("loan_status",axis=1)
y=dataset["loan_status"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

model=GaussianNB()

model.fit(X_train,y_train)

y_pred=model.predict(X_test)
print("accuracy score:",accuracy_score(y_test,y_pred))
print("classification report:",classification_report(y_test,y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Load CSV
dataset = pd.read_csv("/content/breast-cancer.csv")

# Drop unwanted string columns if present
dataset = dataset.drop(columns=["id"], errors="ignore")

# Encode target
X = dataset.drop("diagnosis", axis=1)
y = dataset["diagnosis"].map({"M": 1, "B": 0})  # Encode M=1, B=0

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=43)

# Train model
model = SVC(kernel="linear")
model.fit(X_train, y_train)

# Predict & evaluate
y_pred = model.predict(X_test)
print("classification report:\n", classification_report(y_test, y_pred))
```

```python
# KNN on Iris Dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Load dataset
dataset = pd.read_csv("/content/IRIS.csv")

# Features and target
X = dataset.drop("species", axis=1)   # features
y = dataset["species"]                # target (Setosa, Versicolor, Virginica)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize KNN (let’s start with k=5)
model = KNeighborsClassifier(n_neighbors=3)

# Train model
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
